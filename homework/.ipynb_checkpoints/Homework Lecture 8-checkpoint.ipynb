Write a shooting method to calculate the Legendre polynomials with $n=5$

$$
\frac{d^2p_n}{dx^2} = \frac{1}{1-x^2}[2x\frac{dP_n}{dx} - n(n+1)P_n],
$$
$$
P_n(-1) = (-1)^n \quad P_n(1) = 1
$$


Given the opportunity to reduce the equation to 

$$
y'' = \frac{1}{1-x^2}[2xy' - n(n+1)y],
$$
such that $y(0)=0$, $y'(0) = y'_a$, and $n=1,3,....$.

Vary y'(0) until y(1) =1 with bisection method

```python
import numpy as np
import matplotlib.pyplot as plt
x = 0
x_lim = .9999999
y = np.array([0.0,1.875]) #y - initial condition , dy - arbitrary #
dx = 1.e-4 #Recommended to try 1e-4
n = 2
#Function that we look to integrate
def f(x,y,n):
    f_l = np.zeros(n)
    f_l[0] = y[1]
    f_l[1] = 1/(1-x**2)*(2*x*y[1] - 30*y[0])
    return f_l
    
#Recall rk4
def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y,n)
    k2=h*f(t+0.5*h,y+0.5*k1,n)
    k3=h*f(t+0.5*h,y+0.5*k2,n)
    k4=h*f(t+h,y+k3,n)
    y+=(k1+2.0*k2+2.0*k3+k4)/6.0
    return y

y_r = np.empty(0)
x_r = np.empty(0)
y_o = np.empty(0)
x_o = np.empty(0)
y_l = np.empty(0)
x_l = np.empty(0)

while(x<x_lim):
    if((x+dx)>x_lim):
        dx=x_lim-x
    y = rk4(x,dx,y,n)
    y_r =np.append(y_r,y[0])
    x_r = np.append(x_r, x)
    x+=dx
x = 0
y = np.array([0.0,1.5]) #y - initial condition , dy - arbitrary #
dx = 1.e-4 #Recommended to try 1e-4

while(x<x_lim):
    if((x+dx)>x_lim):
        dx=x_lim-x
    y = rk4(x,dx,y,n)
    y_o =np.append(y_o,y[0])
    x_o = np.append(x_o, x)
    x+=dx
x = 0
y = np.array([0.0,2.5]) #y - initial condition , dy - arbitrary #
dx = 1.e-4 #Recommended to try 1e-4

while(x<x_lim):
    if((x+dx)>x_lim):
        dx=x_lim-x
    y = rk4(x,dx,y,n)
    y_l =np.append(y_l,y[0])
    x_l = np.append(x_l, x)
    x+=dx
plt.plot(x_r, y_r, label = "y_0=1.875")
plt.plot(x_o, y_o, label = "y_0=1.5")
plt.plot(x_l, y_l, label = "y_0=2.5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
